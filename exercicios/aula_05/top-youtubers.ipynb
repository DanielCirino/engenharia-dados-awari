```python
import pandas as pd
```

```python
path_dataset = "./in/topyoutube.csv"
df_youtubers = pd.read_csv(path_dataset)
df_youtubers.info()
```

```python
df_youtubers["100M"] = df_youtubers["100M"].fillna(0.0)
df_youtubers.info()
```

```python
df_youtubers["Total Views"]=df_youtubers["Total Views"].str.replace(",","")
df_youtubers["Total Views"]=df_youtubers["Total Views"].astype("float")
df_youtubers.info()
```

```python
df_youtubers["Total Views"]=df_youtubers["Total Views"]*100000000
df_youtubers
```

```python
df_youtubers=df_youtubers.round({"Avg":2})
df_youtubers
```
